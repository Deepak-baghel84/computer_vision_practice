```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('./image.jpg')

rgb_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10,7))

plt.imshow(rgb_img)

```

```python
#  crop the images and save

x,y,h,w = 15,15,160,225

for columns in range(6):
    x = 15
    if columns > 0:
        y = y+h
    for row in range(3):
        crp_img = rgb_img[y:y+h,x:x+w ]   # for image (height,width)

        x = x+w
        y = y
        cv2.imwrite(f"./result/crop_image.jpg_{columns,row}",crp_img)
    print("\n")
```
